## Limpeza dos dados

Etapa muito importante, devemos sempre trabalhar com dados já tratados, para que os nossos algortimso sejam mais acertivo, neste dataset, existem registros com valores nulos, pedidos cancelados e textos desformatados. Todas estas etapas podem prejudicar nos nossos resultados.

```python
#remover espaços adicionais no campo Descrição do produto
dados['Description'] = dados['Description'].str.strip()
```

```python
#Remover valore Nulos
dados.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
dados.dropna(axis=0, subset=['CustomerID'], inplace=True)
```

```python
#remover os pedidos cancelados
dados['InvoiceNo'] = dados['InvoiceNo'].astype('str')
dados = dados[~dados['InvoiceNo'].str.contains('C')]
```

## Transformação dos dados

Para aplicação do algoritmo Regras de associação é necessário converter a base de dados em uma matriz, que exibe a ocorrência ( foi comprado ou não) de produtos.
Ou seja, primeiros precisamos agrupar esses produtos por pedido e tranforma-los em uma matriz de 0 e 1. Sendo 0 que não houve a compra do produto e 1 que foi efetuada a compra.

```python
cesta_produtos = (dados[dados['Country'] =="France"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
cesta_produtos
```

```python
def transformar_unidades(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
conjunto_cesta_produtos = cesta_produtos.applymap(transformar_unidades)
```

```python
conjunto_cesta_produtos
```

```python

conjunto_cesta_produtos.drop('POSTAGE', inplace=True, axis=1)

```

## Aplicando o Algoritmo

Agora que nossos dados passaram por uma etapa de limpeza e uma etapa de transformação, vamos aplicar o algoritmo de regras de associação. Usando a biblioteca Mlxtend. Estou adotando 90% de confiança e 8% de suporte.


```python
frequencia_itens = apriori(conjunto_cesta_produtos, min_support=0.08, use_colnames=True)

regras = association_rules(frequencia_itens, metric='confidence', min_threshold=0.9)
regras
```

Então, como podemos ver, pratos de plásticos vermelhos são comprados junto com copos de plastico vermelho, provavelmente para festas. Isso nós da um insign bacana,podemos aumentar a populariedade de pratos e recomentar os copos, a fim de influenciar as compras casadas

## Comparativo com o Reino Unido

```python
cesta_produtos_reino_unidos = (dados[dados['Country'] =="United Kingdom"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
conjunto_cesta_produtos_reino_unido = cesta_produtos_reino_unidos.applymap(transformar_unidades)
conjunto_cesta_produtos_reino_unido.drop('POSTAGE', inplace=True, axis=1)
```

```python
frequencia_itens_reino_unido = apriori(conjunto_cesta_produtos_reino_unido, min_support=0.04, use_colnames=True)

regras_reino_unido = association_rules(frequencia_itens_reino_unido, metric='confidence', min_threshold=0.7)
regras_reino_unido
```

Então podemos ver que neste exemplo, as regras nos retornou que os padrões de compra dos ingleses são xicaras  de chás. Ou seja, influenciando na compra casada de chicarás de diferentes cores e estampas há maior probalidade de compras! Este insights para alguns pode parecer óbvio, mas para um especialista que de antemão não conhece a cultura do país Inglês, pode se surprender e ter resultados interessantes.

## Comparativo com a Alemanha

```python
cesta_produtos_alemanha = (dados[dados['Country'] =="Germany"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
conjunto_cesta_produtos_alemanha = cesta_produtos_alemanha.applymap(transformar_unidades)
conjunto_cesta_produtos_alemanha.drop('POSTAGE', inplace=True, axis=1)
```

```python
frequencia_itens_alemanha = apriori(conjunto_cesta_produtos_alemanha, min_support=0.05, use_colnames=True)

regras_alemanha = association_rules(frequencia_itens_alemanha, metric='confidence', min_threshold=0.8)
regras_alemanha
```

Como podemos ver, os padrões de compras giram em torno de Bolsas que podem ser utilziada para levar comida, assim como recipientes em forma de frutas, seria interessante investiar mais a respeito. Mas com  este resultado, poderia ser feito um impulsionamento nas ofertas sobre bolsas desta categoria e recomendar em conjunto os recipientes.
Estas regras apresentam até 13% de suporte e 84% de confiança. tendo um lift maior que 1, que nos diz que a compra do consequente, na maioria das vezes, ocorre com estes antecentes que foram mostrados.
